```python
df = pd.DataFrame(pd.read_csv('/content/NIFTY50_all.csv'))
df
```

```python
array = df.Symbol.unique()
label = []
```

```python
for i in range(len(array)):
  label.append(len(df[df['Symbol']==array[i]]))
```

```python
plt.figure(figsize=(15,10))
plt.pie(label,labels=array, autopct='%1.1f%%')
plt.show()
```

```python
#Bar Chart
plt.bar(df['Symbol'],df['Open'])
plt.show()
```

### **Histogram**
Similar to bar plot but is plotted with respect to a continous variable; hepls us to find the concentration of values and the difference between the variable

```python
plt.figure(figsize=(15,7))
plt.hist(df['Symbol'],facecolor = 'Yellow', edgecolor='#169acf', linewidth=0.5)
plt.show()
```

### **Gantt Chart**
Project timeline or progression of works; time intervals on the horizontal axis and works is listed on vertical axis

```python
listOf_Xticks = np.arange(0, 25000, 1000)
plt.xticks(listOf_Xticks)
plt.barh(df['Symbol'], df['Open'], left=df['Close'],color='green')
plt.show()
```

### Heat Map
Visualize the difference between the data shown by the gradient of colors; Can be used to show correlation between two variables and helps to identify it ver easily.

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.corr("pearson"),annot = True)
plt.show()
```

```python
data=[df['Low'],df['High']]

plot=plt.boxplot(data)
plt.show()
```

```python
import plotly.express as px
fig=px.area(df['Low'],df['High'],color=df['Symbol'])
fig.show()
```

```python
plt.scatter(df['Date'],df['Low'],color='Black',s=0.2)
plt.show()
```

```python
import plotly.graph_objects as go
fig = go.Figure(go.Waterfall(orientation = "v", x=df['Symbol'],textposition = "outside",y = df['Open'],
    connector = {"line":{"color":"rgb(63, 63, 63)"}},
))

fig.update_layout(
        title = "Stock Market Opening values",
        showlegend = True
)

fig.show()
```

```python
import plotly.graph_objects as go

candlestick = go.Candlestick(
                            x=df.index,
                            open=df['Open'],
                            high=df['High'],
                            low=df['Low'],
                            close=df['Close']
                            )

fig = go.Figure(data=[candlestick])

fig.show()
```
